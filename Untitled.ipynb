```python
import pandas as pd
df = pd.read_csv("customer_shopping_behavior.csv")
                 
```

```python
df.head()
```

```python
df.describe()
#It describes only numerical columns
#if we need summarical statistics of all columns, we can use (include=all)
```

```python
df.isnull().sum()
#to know the null values in the columns
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower() # here we apply lowercase to the column names
df.columns = df.columns.str.replace(' ','_') #replaced Spaces between the words and replaced it to under score
df.columns # checked for the changes we made
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) # here we replaed the name of the column
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
#Here we split the customers in 4 groups based on their ages, 
#here q stands for quantile and qcut cuts the data in an equal sized groups
#created new column named "age_group" and insert the cutted data with the labels in it based on the customer's age's 
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = { 'Fortnightly': 14,
                     'Weekly': 7,
                     'Bi-Weekly': 14,
                     'Monthly': 30,
                     'Annually': 365,
                     'Every 3 Months': 90 } 
df ['purchase_frequency_days'] = df ['frequency_of_purchases'].map(frequency_mapping) #Here created column named as purchase_frequency_days
```

```python
df [['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df [['discount_applied', 'promo_code_used']].head(10) # Seen that discount_applied and promo_code_used columns have same data
```

```python
(df['discount_applied'] == df['promo_code_used']).all() # Both the columns have same data as we can see
```

```python
df = df.drop('promo_code_used', axis=1) 
```
